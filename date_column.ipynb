```python
#drop useless rows
data.drop(labels=[0, 1, 824], axis=0, inplace=True)
```

```python
# Make correct Date and ID column names
data.rename(columns={"Unnamed: 0":"Order_id", "Ship Mode":"Order_date"}, inplace=True)
```

```python
# Reset Index
data.reset_index(drop=True, inplace=True)
```

```python
# Create payment column
data["payment"]=np.nan
```

```python
# function to fill the payment column through np.where
def pay(x):
  return np.where(~x.isnull(), x, data.payment)
```

```python
#First Class
data["payment"]=pay(data["First Class"])
data["payment"]=pay(data["Unnamed: 3"])
data["payment"]=pay(data["Unnamed: 4"])
```

```python
# check null values to see everything goes right
data.payment.isnull().sum()
```

```python
# Create segment column
data["segment"]=np.nan
```

```python
# fill segment column accordingly First Class 's segments
data["segment"]=np.where(~data["First Class"].isnull(), "Consumer", data.segment)
data["segment"]=np.where(~data["Unnamed: 3"].isnull(), "Corporate", data.segment)
data["segment"]=np.where(~data["Unnamed: 4"].isnull(), "Home Office", data.segment)
```

```python
# Create ship_mode column
data["ship_mode"]=np.nan
```

```python
#fill ship_mode column with First Class
data["ship_mode"]=np.where(~data.payment.isnull(), "First Class", data.ship_mode)
```

```python
# fill payment column through Same day
data["payment"]=pay(data["Same Day"])
data["payment"]=pay(data["Unnamed: 6"])
data["payment"]=pay(data["Unnamed: 7"])
```

```python
data.payment.isnull().sum()
```

```python
#fill ship_mode column with Same Day
data["ship_mode"]=np.where(~data.payment.isnull() & data.segment.str.contains('nan'), "Same Day", data.ship_mode)
```

```python
#fill ship_mode column accordingly Same Day's segments
data["segment"]=np.where(~data["Same Day"].isnull(), "Consumer", data.segment)
data["segment"]=np.where(~data["Unnamed: 6"].isnull(), "Corporate", data.segment)
data["segment"]=np.where(~data["Unnamed: 7"].isnull(), "Home Office", data.segment)
```

```python
#fill payment column through Second Class
data["payment"]=pay(data["Second Class"])
data["payment"]=pay(data["Unnamed: 9"])
data["payment"]=pay(data["Unnamed: 10"])
```

```python
data.payment.isnull().sum()
```

```python
#fill ship_mode column with Second Class
data["ship_mode"]=np.where(~data.payment.isnull() & data.segment.str.contains('nan'), "Second Class", data.ship_mode)
```

```python
#fill ship_mode column accordingly Second Class's segments
data["segment"]=np.where(~data["Second Class"].isnull(), "Consumer", data.segment)
data["segment"]=np.where(~data["Unnamed: 9"].isnull(), "Corporate", data.segment)
data["segment"]=np.where(~data["Unnamed: 10"].isnull(), "Home Office", data.segment)
```

```python
#fill payment column through Standard Class
data["payment"]=pay(data["Standard Class"])
data["payment"]=pay(data["Unnamed: 12"])
data["payment"]=pay(data["Unnamed: 13"])
```

```python
data.payment.isnull().sum()
```

```python
#fill ship_mode column with Standard Class
data["ship_mode"]=np.where(~data.payment.isnull() & data.segment.str.contains('nan'), "Standard Class", data.ship_mode)
```

```python
#fill ship_mode column accordingly Standard Class's segments
data["segment"]=np.where(~data["Standard Class"].isnull(), "Consumer", data.segment)
data["segment"]=np.where(~data["Unnamed: 12"].isnull(), "Corporate", data.segment)
data["segment"]=np.where(~data["Unnamed: 13"].isnull(), "Home Office", data.segment)
```
